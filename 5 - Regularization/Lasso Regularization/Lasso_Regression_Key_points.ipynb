# **1. How are coefficients affected?**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import Lasso
from sklearn.metrics import r2_score
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
data = load_diabetes()
df = pd.DataFrame(data.data,columns=data.feature_names)
df['TARGET'] = data.target
df.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(data.data,data.target,test_size=0.2,random_state=2)
```

```python
coefs = []
r2_scores = []

for i in [0,0.1,1,10]:
    reg = Lasso(alpha=i)
    reg.fit(X_train,y_train)
    
    coefs.append(reg.coef_.tolist())
    y_pred = reg.predict(X_test)
    r2_scores.append(r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(14,9))
plt.subplot(221)
plt.bar(data.feature_names,coefs[0])
plt.title('Alpha = 0 ,r2_score = {}'.format(round(r2_scores[0],2)))

plt.subplot(222)
plt.bar(data.feature_names,coefs[1])
plt.title('Alpha = 0.1 ,r2_score = {}'.format(round(r2_scores[1],2)))

plt.subplot(223)
plt.bar(data.feature_names,coefs[2])
plt.title('Alpha = 1 ,r2_score = {}'.format(round(r2_scores[2],2)))

plt.subplot(224)
plt.bar(data.feature_names,coefs[3])
plt.title('Alpha = 10 ,r2_score = {}'.format(round(r2_scores[3],2)))

plt.show()
```

As you increase the alpha value the coefficients decreases and becomes 0.

# **2. Higher Coefficients are affected more**

```python
alphas = [0.0001,0.001,0.01,0.1,1,10,100,1000,10000]
coefs = []

for i in alphas:
    reg = Lasso(alpha=i)
    reg.fit(X_train,y_train)
    
    coefs.append(reg.coef_.tolist())
```

```python
input_array = np.array(coefs)

coef_df = pd.DataFrame(input_array,columns=data.feature_names)
coef_df['alpha'] = alphas
coef_df.set_index('alpha')
```

```python
alphas = [0.0001,0.0005,0.001,0.005,0.1,0.5,1,5,10]

coefs = []

for i in alphas:
    reg = Lasso(alpha=i)
    reg.fit(X_train,y_train)
    
    coefs.append(reg.coef_.tolist())
```

```python
input_array = np.array(coefs).T

plt.figure(figsize=(15,8))
plt.plot(alphas,np.zeros(len(alphas)),color='black',linewidth=5)
for i in range(input_array.shape[0]):
    plt.plot(alphas,input_array[i],label=data.feature_names[i], marker = 'o', markersize=10)
plt.legend()
```

The rate of change of higher coefficients (-ve or +ve) is more than the smaller, but the smaller ones becomes 0 first

# **3. Impact on Bias and Variance**

```python
m = 100
X = 5 * np.random.rand(m, 1) - 2
y = 0.7 * X ** 2 - 2 * X + 3 + np.random.randn(m, 1)

plt.scatter(X, y)
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.reshape(100,1),y.reshape(100),test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)

X_train = poly.fit_transform(X_train)
X_test = poly.transform(X_test)
```

```python
from mlxtend.evaluate import bias_variance_decomp

alphas = np.linspace(0,30,100)

loss = []
bias = []
variance = []

for i in alphas:
    reg = Lasso(alpha=i)
    avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        reg, X_train, y_train, X_test, y_test, 
        loss='mse',
        random_seed=123)
    loss.append(avg_expected_loss)
    bias.append(avg_bias)
    variance.append(avg_var)
```

```python
plt.plot(alphas,loss,label='loss')
plt.plot(alphas,bias,label='Bias')
plt.plot(alphas,variance,label='Variance')
plt.xlabel('Alpha')
plt.legend()
```

# **4. Effect of Regularization on Loss Function**

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)

plt.scatter(X,y)

reg = LinearRegression()
reg.fit(X,y)

print(reg.coef_)
print(reg.intercept_)
```

```python
def cal_loss(m,alpha):
    return np.sum((y - m*X.ravel() + 2.29)**2) + (alpha * np.abs(m))
    
def predict(m):
    return m*X - 2.29
```

```python
m = np.linspace(-45,100,100)
plt.figure(figsize=(7,7))
for j in [0,100,500,1000,2500,3500,4500,5500]:
    loss = []
    for i in range(m.shape[0]):
        loss_i = cal_loss(m[i],j)
        loss.append(loss_i)
    plt.plot(m,loss,label='alpha = {}'.format(j))
plt.legend()
plt.xlabel('weight')
plt.ylabel('Loss')
plt.show()
```

1. Contour Lines of the Cost Function:

- When  is small, the contour lines of the cost function are mostly determined by the error term, which forms a smooth bowl or ellipse shape (because of square terms).
- As  increases, the influence of the L1 penalty term     becomes stronger, and the contours of the cost function start to be pulled into a diamond shape.

2. Cost Function Minimization:

- At higher , the optimization process moves from the elliptical contours to the diamond-shaped contours of the L1 penalty. 
- The L1 penalty term introduces a sharp corner or "kink" at  = 0. 

3. Resulting Angle: 

- The diamond shape of the penalty term creates a situation where the optimization path has to align with the corners of the diamond. 
- When you increase , the constraints created by the L1 penalty force the coefficients to be exactly at the corners of the diamond (i.e.,   = 0 or non-zero non-zero). 

4. Why Only at  = 0:
- The penalty term     is non-differentiable at  = 0, which means that the optimization path creates an angle where the best choice for minimizing the cost function can be exactly at  = 0 or on the boundary defined by the diamond.
